# Exercise 1: Running PCA on simulated spike traces.

In this exercise, you'll analyze a dataset of simulated spike traces, loaded in variable `neuron_spike_hist` below.

````python
from scipy.io import loadmat
import sklearn.decomposition
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

neuron_spike_hist = np.load('neuron_spike_hist.npy')
plt.imshow(neuron_spike_hist, cmap='Greens', aspect='auto')
plt.ylabel('Neurons')
plt.xlabel('Time');
````

**Task**: Do PCA on this dataset, finding the PCs to reduce the dimensionality of the neurons. Take a careful look at the dimensions after PCA to make sure you've reduced the dimensionality across the correct axis. If directly computing eigenvalues using `np.linalg.eig`, make sure to center your data and take the eigendecomposition of the covariance matrix of your data.

**Question**: To validate your PCA, plot your eigenvector with highest eigenvalue. Does this look like a major time-series motif? If it doesn't, double check your code.

**Question**: After plotting the explained variance ratio, how many PCs do you think are needed to represent this dataset?

````python
# ANSWER
pca = sklearn.decomposition.PCA()
pca.fit(neuron_spike_hist)

plt.figure()
plt.plot(pca.components_[np.argmax(pca.explained_variance_)])
plt.title('Top eigenvector')

plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('# components')
plt.ylabel('explained variance fraction')
````

````python
# ANSWER

centered_nsh = neuron_spike_hist - neuron_spike_hist.mean(axis=0, keepdims=True)
cov = centered_nsh.T@centered_nsh / (centered_nsh.shape[0]-1) # writing it out by hand
cov = np.cov(centered_nsh.T) # alternatively, use numpy
evals, evecs = np.linalg.eig(cov)

plt.figure()
plt.plot(evecs[:, np.argmax(evals)])
plt.title('Top eigenvector')

exp_var_ratio = np.abs(evals) / np.abs(evals).sum()
plt.figure()
plt.plot(np.cumsum(exp_var_ratio))
plt.xlabel('# components')
plt.ylabel('explained variance fraction')
````

````python
# ANSWER
# compare our two ways of computing
print('comparing sklearn & ours')
print(pca.explained_variance_[:10])
print(evals[:10])
````

````python
from scipy.io import loadmat
import sklearn.decomposition
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
````

````python
data = loadmat('SMA_sampleData.mat', squeeze_me=True)['D_sma']
print('SMA keys:', data.dtype)

neural_data = data['A'][0]
print('neural data shape', neural_data.shape)
````

Now we can do PCA! Refer back to lecture to choose a way to implement this. Think carefully about what dimensionality of your input to PCA; We want to preserve the time dimension (size 5330), while reducing the dimensionality of the nueral dimension (size 77). If you're using `sklearn`, check out the documentation for `PCA.fit()` [here](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html#sklearn.decomposition.PCA.fit), you'll have to think about what our "samples" and "features" are.
````python
# ANSWER
pca = sklearn.decomposition.PCA()
pca.fit(data['A'][0].T)
print(pca.components_.shape)
````

**Question**: What do the top 3 PCs seem to represent?

````python
# ANSWER
from matplotlib.collections import LineCollection

x = pca.components_[1]
y = pca.components_[2]

points = np.array([x, y]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)

lc = LineCollection(segments, cmap='Reds')
lc.set_array(np.arange(pca.components_.shape[1]))
f, ax = plt.subplots()
ax.add_collection(lc)
plt.xlim([x.min(), x.max()])
plt.ylim([y.min(), y.max()]);
````

````python
# ANSWER
from mpl_toolkits.mplot3d.art3d import Line3DCollection

pcs = pca.components_[:3].T
segments = np.stack([pcs[1:], pcs[:-1]], axis=1)
lc = Line3DCollection(segments, cmap='viridis')
lc.set_array(np.arange(pca.components_.shape[1])) # For colors

f = plt.figure()
ax = f.add_subplot(111, projection='3d')
ax.add_collection(lc)
ax.view_init(elev=0, azim=220)
ax.set(
    xlim=[pcs[:, 0].min(), pcs[:, 0].max()],
    ylim=[pcs[:, 1].min(), pcs[:, 1].max()],
    zlim=[pcs[:, 2].min(), pcs[:, 2].max()],
);
````
